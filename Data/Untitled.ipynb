```python
import pandas as pd

od = pd.read_csv('od_final.csv')
od = od.drop(['Prev_Spatial_Mean', 'Prev_Spatial_Max', 'Pct_Majority_Income_to_Housing', 'Premature_Deaths', 'Spatial_Max'], axis = 1)
od
```

```python
import missingno as msno

msno.matrix(od)
```

```python
preplot = od['Overdose_Rate_per_100k'].hist()
preplot.set_xlabel("Drug Overdose Rate per 100k")
preplot.set_ylabel("Frequency")
preplot.set_title('Histogram of Overdose_Rate_per_100k')
```

```python
import numpy as np
postplot = np.log(od['Overdose_Rate_per_100k']).hist()
postplot.set_xlabel("Drug Overdose Rate per 100k")
postplot.set_ylabel("Frequency")
postplot.set_title('Histogram of log_Overdose_Rate_per_100k')
```
